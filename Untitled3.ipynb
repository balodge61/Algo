```python
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
stock_list = ["MLR","HCA","LPX","JPM","F","LYB"]
period = "60d"
interval = "2m"
stock1 = yf.Ticker("SPY")
stock1_price = stock1.history(period=period, interval=interval)
stock1_df = pd.DataFrame(stock1_price.Close)
df = stock1_df

for stock in stock_list:

    stock2 = yf.Ticker(stock)
    stock2_price = stock2.history(period=period, interval=interval)
    stock2_df = pd.DataFrame(stock2_price.Close)
    df = df.merge(stock2_df, left_on='Datetime', right_on='Datetime')
    
df.head()
df.corr()
```

```python
period = "60d"
interval = "5m"
stock = yf.Ticker("JPM")
df = stock.history(period=period, interval=interval)

df = df.replace(to_replace = 0, value =1) 
df = df.drop(['Dividends','Stock Splits'], axis = 1)
df = df.pct_change()*100

df["Volume"] = df.Volume
df["Close"] = df.Close
df["1p_Prior"] = df.Close.shift(1)
df["2p_Prior"] = df.Close.shift(2)
df["1p_Future"] = df.Close.shift(-1)

df.head()
```

```python
df = df.dropna() 
print(df.shape)
df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

model = LinearRegression()


X = df[['Open','High','Low','Close','Volume','1p_Prior','2p_Prior']]
y = df[['1p_Future']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model.fit(X_train, y_train)

predictions = model.predict(X_test)
MSE = mean_squared_error(y_test, predictions)
r2 = model.score(X_test, y_test)

print(f"MSE: {MSE}, R2: {r2}")
print(predictions)
print(model.coef_)
```
