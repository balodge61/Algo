```python
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
stock_list = ["MLR","HCA","LPX", "KO", "PEP", "GS", "JPM"]
stock = "SPY"
period = "60d"
interval = "30m"
stock1 = yf.Ticker(stock)
stock1_price = stock1.history(period=period, interval=interval)
stock1_df = pd.DataFrame(stock1_price.Close).rename(columns={"Close": stock})
df = stock1_df

for stock in stock_list:
    print("Starting stock: ", stock)
    stock2 = yf.Ticker(stock)
    print("Got stock: ", stock)
    stock2_price = stock2.history(period=period, interval=interval)
    stock2_df = pd.DataFrame(stock2_price.Close).rename(columns={"Close": stock})
    df = df.merge(stock2_df, left_on='Datetime', right_on='Datetime')
    
df.head()
```

```python
period = "60d"
interval = "5m"
stock = yf.Ticker("JPM")
df = stock.history(period=period, interval=interval)

df = df.replace(to_replace = 0, value =1) 
df = df.drop(['Dividends','Stock Splits'], axis = 1)

cols_pct_change = ["Close", "Volume"]
col_names = []

def target(future_close_pc):
    if future_close_pc >= .001:
        return 1
    else:
        return 0

for col in cols_pct_change:
    for i in np.arange(1,5):
        col_name = f"{col}_{i}p_pc"
        df[col_name] = df[col].pct_change(periods=i)*100
        col_names.append(col_name)
    for i in np.arange(-4, 0):
        if col == "Close":
            df[f"{col}_f{-i}p_pc"] = df[col].pct_change(periods=i)*100
            df[f"y{-i}"] = df[col].pct_change(periods=i).apply(target)

df = df.dropna() 
print(df.shape)
df.head()
```

```python
KO= yf.Ticker("GME")
KO.option_chain("2021-03-05")
```

```python
#BROKE AF
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

model = LinearRegression()


X = df[['Open','High','Low','Close','Volume'] + col_names]
y = df['Close_f1p_pc']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model.fit(X_train, y_train)

predictions = model.predict(X_test)
MSE = mean_squared_error(y_test, predictions)
r2 = model.score(X_test, y_test)

print(f"MSE: {MSE}, R2: {r2}")
print(predictions)
print(model.coef_)
```
